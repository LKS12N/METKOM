```python
#Tugas 1 Hukum inersia
def hukum_inersia():
    print("Hukum Newton 1: Hukum Inersia")

    massa = float(input("Masukkan massa benda (kg): "))
    kecepatan = float(input("Masukkan kecepatan benda (m/s): "))

    if kecepatan == 0:
        print("Benda diam")
    else:
        print("Benda bergerak lurus dengan kecepatan konstan")

hukum_inersia()
```

```python
#Tugas 1 Hukum Percepatan
def hukum_gaya_percepatan():
    print("Hukum Newton 2: Hukum Gaya dan Percepatan")

    massa = float(input("Masukkan massa benda (kg): "))
    gaya = float(input("Masukkan gaya yang dikenakan (N): "))

    percepatan = gaya / massa

    print(f"Percepatan benda={percepatan} m/s^2")

hukum_gaya_percepatan()
```

```python
#Tugas 1 Hukum Aksi dan Reaksi
def hukum_aksi_reaksi():
    print("Hukum Newton 3: Hukum Aksi dan Reaksi")

    gaya = float(input("Masukkan gaya yang dikenakan (N): "))

    reaksi = -gaya

    print(f"Benda bergerak dengan gaya {gaya} N dan reaksi {reaksi} N")

hukum_aksi_reaksi()
```

```python
#Tugas 2
def pilih_tempat_makan(lapar, keinginan):
    if lapar:
        if keinginan == "yg cepat":
            return "Warmindo banyak."
        elif keinginan == "sehat":
            return "Makan dimana ada yg penting ada sayur."
        elif keinginan == "berat":
            return "mau makan seberat apa sih."
    else:
        if keinginan == "santai":
            return "Warmindo lagi lah."
        elif keinginan == "sama temen":
            return "Kemana lagi kalo ga ke warmindo."
        else:
            return "Pilih yg pasti aja."

#Input dari pengguna
lapar_input = input("Laper ga sih? (ya/tidak): ").strip().lower()
lapar = lapar_input == "ya"

keinginan = input("Maunya apa? (yg cepat/sehat/berat/santai/sama temen): ").strip().lower()

#Hasil
print(pilih_tempat_makan(lapar, keinginan))
```
